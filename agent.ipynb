```python
from dotenv import load_dotenv
load_dotenv()
```

```python

from langchain.chat_models import init_chat_model
from langchain.messages import HumanMessage, AIMessage, SystemMessage
from langchain_google_genai import ChatGoogleGenerativeAI

context_model =  ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0,
    max_tokens=10,
    timeout=None,
    max_retries=2,
    # other params...
)

chat_model =  ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0,
    max_tokens=None,
    timeout=None,
    max_retries=2,
    # other params...
)

context_message = SystemMessage(content="Give a 3 word topic name to given  conversation."),

system_msg = SystemMessage("You are a helpful assistant.")
human_msg = HumanMessage("Hello, how are you?")

messages = [system_msg, human_msg]

while True:
    user_input = input("User: ")

    if user_input.lower() in ["exit", "quit"]:
        break
    hu = HumanMessage(content=user_input)
    messages.append(hu)
    response = chat_model.invoke(user_input)
    ai = AIMessage(content=response.content)

    messages.append(ai)
    current_massage = [context_message, hu, ai]
    context_response = context_model.invoke(str(current_massage))

    print("AI:", response.content)
    print("Context Topic:", context_response.content)
```
